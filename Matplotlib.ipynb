# Plotting With Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. Simple plot with numbers

```python
plt.plot([1, 3, 2, 5, 4])
plt.show()
```

### 2. Points with X and Y coordinates

```python
xValues = [1, 2, 3, 4]
yValues = [1, 4, 9, 16]
plt.plot(xValues, yValues)
plt.title("Test Plot", fontsize=14, color="r")
plt.xlabel("x", fontsize=12, color="g")
plt.ylabel("y = x^2", fontsize=12, color="b")
plt.show()
```

### 3. Change figure size

```python
plt.figure(figsize=(2, 5)) # 2 inches wide, 5 inches high
plt.plot(xValues, yValues, "ro") # red-O
plt.axis([0, 6, 0, 20]) # x-min, x-max, y-min, y-max
plt.annotate("square it", (3, 6))
plt.show()
```

### 5. Two sets of 10 random dots

```python
d = {"Red O": np.random.rand(10),
     "Grn X": np.random.rand(10)}
df = pd.DataFrame(d)

df.plot(style=["ro", "gx"])
plt.show()
```

```python
# https://github.com/joeyajames/Python/blob/master/Pandas/Fremont_weather.txt
df = pd.read_csv("Fremont_weather.csv")
print(df)
plt.bar(df["month"], df["record_high"]["record_low"], color="r")
plt.bar(df["month"], df["record_low"], color="c")
plt.plot(df["month"], df["avg_high"], color="k")
plt.plot(df["month"], df["avg_low"], color="b")
plt.legend()
plt.show()
```

### 6. Multiple plots in single figure

```python
fig = plt.figure()

fig.add_subplot(221)
plt.plot([5, 4, 8, 3])

fig.add_subplot(222)
plt.bar([1, 2, 3, 4], [8, 4, 7, 3])

fig.add_subplot(223)
plt.scatter([1, 2, 3, 4], [8, 4, 7, 3])

fig.add_subplot(224)
plt.pie([7, 4, 9, 2, 5])

plt.show()
```
